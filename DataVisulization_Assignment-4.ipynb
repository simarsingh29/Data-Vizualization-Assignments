```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Create a random dataset of 100 rows × 30 columns with values in [1,200]
np.random.seed(42)
data = np.random.randint(1, 201, size=(100, 30))
df = pd.DataFrame(data, columns=[f"Col_{i+1}" for i in range(30)])

print("Initial Dataset (first 5 rows):")
display(df.head())

# 1(a).
df_na = df.mask(df.applymap(lambda x: 10 <= x <= 60))

# Count of NaN in each row and column
row_na_count = df_na.isna().sum(axis=1)
col_na_count = df_na.isna().sum(axis=0)

print("\nCount of NaN values in each row (first 10 rows):")
display(row_na_count.head(10))

print("\nCount of NaN values in each column:")
display(col_na_count)

# 2. Replace NaN with column mean
df_filled = df_na.apply(lambda col: col.fillna(col.mean()), axis=0)

print("\nDataset after filling NaN with column means (first 5 rows):")
display(df_filled.head())

# 3. Plot Heatmap (missing values)
plt.figure(figsize=(12, 6))
sns.heatmap(df_na.isna(), cbar=False, cmap="viridis")
plt.title("Heatmap of Missing Values (Before Replacement)")
plt.show()

# 4. Count number of columns with correlation <= 0.7
corr_matrix = df_filled.corr()
col_corr_counts = (corr_matrix.abs() <= 0.7).sum()

print("\nNumber of correlations <= 0.7 for each column:")
display(col_corr_counts)

# 5. Normalize each column to [0,10]
df_normalized = (df_filled - df_filled.min()) / (df_filled.max() - df_filled.min()) * 10

print("\nNormalized Dataset (first 5 rows):")
display(df_normalized.head())

# 6. Replace values: <=5 → 0, >5 → 1
df_binary = df_normalized.applymap(lambda x: 0 if x <= 5 else 1)

print("\nBinary Dataset (first 5 rows):")
display(df_binary.head())

# 7. Visualizing distribution of dataset (practice seaborn distributions)
plt.figure(figsize=(10, 5))
sns.histplot(df_filled["Col_1"], kde=True, bins=20, color="blue")
plt.title("Distribution of Col_1 with KDE")
plt.show()

plt.figure(figsize=(10, 5))
sns.kdeplot(df_filled["Col_2"], shade=True, color="red")
plt.title("KDE Plot of Col_2")
plt.show()

plt.figure(figsize=(10, 5))
sns.boxplot(x=df_filled["Col_3"])
plt.title("Boxplot of Col_3")
plt.show()

# Save dataset to CSV for download
df_binary.to_csv("assignment4_dataset.csv", index=False)
print("\n✅ Dataset saved as 'assignment4_dataset.csv'")
```
