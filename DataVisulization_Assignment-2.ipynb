```python
# 8. Create a sample dataset (employees.csv) containing information about employees in a company. 
data = {
    "Employee_ID": [101, 102, 103, 104, 105],
    "Name": ["Alice", "Bob", "Charlie", "Diana", "Edward"],
    "Department": ["HR", "IT", "IT", "Marketing", "Sales"],
    "Age": [29, 34, 41, 28, 38],
    "Salary": [50000, 70000, 65000, 55000, 60000],
    "Years_of_Experience": [4, 8, 10, 3, 12],
    "Joining_Date": ["2020-03-15", "2017-07-19", "2013-06-01", "2021-02-10", "2010-11-25"],
    "Gender": ["Female", "Male", "Male", "Female", "Male"],
    "Bonus": [5000, 7000, 6000, 4500, 5000],
    "Rating": [4.5, 4.0, 3.8, 4.7, 3.5]
}

df = pd.DataFrame(data)
df
```

```python
# Save to CSV
df.to_csv("employees.csv", index=False)
```

```python
# a) Shape of the DataFrame
df.shape
```

```python
# b) Summary (info)
df.info()
```

```python
# c) Descriptive statistics
df.describe(include="all")
```

```python
# d) First 5 and Last 3 rows
print("\nFirst 5 rows:")
print(df.head())
print("\nLast 3 rows:")
print(df.tail(3))
```

```python
# e) Calculate statistics
avg_salary = df["Salary"].mean()
total_bonus = df["Bonus"].sum()
youngest_age = df["Age"].min()
highest_rating = df["Rating"].max()

print("\nAverage Salary:", avg_salary)
print("Total Bonus:", total_bonus)
print("Youngest Age:", youngest_age)
print("Highest Rating:", highest_rating)
```

```python
# f) Sort by Salary (descending)
df.sort_values(by="Salary", ascending=False)
```

```python
# g) Categorize performance rating
def categorize(rating):
    if rating >= 4.5:
        return "Excellent"
    elif rating >= 4.0:
        return "Good"
    else:
        return "Average"

df["Performance"] = df["Rating"].apply(categorize)
df
```

```python
# h) Missing values
df.isnull().sum()
```

```python
# i) Rename Employee_ID to ID
df.rename(columns={"Employee_ID": "ID"}, inplace=True)
df.head()
```

```python
# j) Filter employees
exp_gt_5 = df[df["Years_of_Experience"] > 5]
it_dept = df[df["Department"] == "IT"]

exp_gt_5
it_dept
```

```python
# k) Add Tax column (10% of Salary)

df["Tax"] = df["Salary"] * 0.10
print("\nWith Tax Column:\n", df)

# l) Save modified DataFrame
df.to_csv("employees_modified.csv", index=False)
print("\nModified DataFrame saved as 'employees_modified.csv'")
```
